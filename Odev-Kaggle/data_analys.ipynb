```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("HousingData.csv")
data
```

```python
data.head(50)
```

```python
# ZN Columunu kaldırsak bence bir şey değişmez. Çünkü çoğu değer 0'a eşit.
```

```python
Zn = data["ZN"] == 0
Zn.value_counts() # ZN columunda 0 a eşit olanların sayısı.
```

```python
data2 = data.drop("ZN" , axis = 1)
data2
```

```python
data2.isnull().sum()
```

```python
# CRIM Sütunu
data2["CRIM"].fillna(data2["CRIM"].mean(), inplace = True)
data2
```

```python
data2.isnull().sum()
```

```python
# İndusları mod ile dolduralım çünkü tekrar ediyor hepsi.
data2["INDUS"].fillna(data2["INDUS"].mode()[0], inplace = True)
data2
```

```python
data2.isnull().sum()
```

```python
# Chas column(u 
(data2["CHAS"] == 0).value_counts() # 452 tane 0 varmış 54 tane yok. Bu columnu kaldırabiliriz.
```

```python
data2 = data2.drop("CHAS", axis = 1)
data2
```

```python
data2.isnull().sum()
```

```python
# AGE mean ile doldurulur
data2["AGE"].fillna(data2["AGE"].mean(),inplace = True)
data2.isnull().any()
```

```python
# LSTAT median ile dolduralım.
data2["LSTAT"].fillna(data2["LSTAT"].median(), inplace = True)
data2
```

```python
data2.isnull().sum()
```

```python
plt.plot(data2["MEDV"],data2["NOX"],marker = "s",color = "red")
plt.legend()
plt.xlabel("MEDV")
plt.ylabel("NOX")
plt.grid()
plt.show()
```

```python
plt.plot(data2["RM"],data2["CRIM"],marker = "x",color = "green")
plt.legend()
plt.xlabel("RM")
plt.ylabel("CRIM")
plt.grid()
plt.show()
```
