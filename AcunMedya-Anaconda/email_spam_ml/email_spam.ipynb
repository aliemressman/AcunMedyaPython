```python
import nltk
nltk.download('stopwords')
```

```python
# StopWord Kaldırma İşlemi. -> the, at, on, is
# Stem işlemi (Kelimeleri köküne indirme) -> Running -> Run
# rUnning - RUNNING - RuNNing

import string
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
# Python DERSLERİMİZ Hakkında. -> "python derslerimiz hakkında" -> ["python","derslerimiz","hakkında"]

ps = PorterStemmer()


def clean_text(text):
    text = text.lower() #Tüm metni küçük harfe çevirdi

    text = text.translate(str.maketrans('','',string.punctuation)) #Tüm noktalama işaretlerini kaldırdı.

    words = text.split() # Kelime kelime ayırıyorum.

    words = [word for word in words if not word.isdigit()]

    words = [ps.stem(word) for word in words if word not in stopwords.words('english')]

    return " ".join(words)

clean_text("1. PythOn Derslerimiz., Hakında 30")
```

```python
df.head(3)
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

X = df["clean_text"]
y = df["label_num"] 

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df["clean_text"])

X_train, X_test, y_train, y_test = train_test_split(X_tfidf,y,test_size = 0.33, random_state = 42)
```

```python
# Naive Bayes
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train,y_train)
```

```python
# Bu modelin metriklerini nasıl ölçüyoruz?
# Bu modeli siz Geliştiriyor olsaydınız, hangi değişiklikler ile modeli performanısın geliştirmek için hangi adımları atardınız.
# KNN algoritmasını araştırıp, aynı modeli KNN ile geliştirelim.
```

```python
# Classification report
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test,y_pred))

# support -> Veri sayısı
# precision -> kesinlik -> doğru şekilde pozitif tahmin yapma yeteneği?
# precision 1.00 -> %100 spam dediklerim spammış. 0.89 -> %11 spam olan maile spam değil demişim.

# recall -> duyarlılık -> 0.68 -> %32 veriyi kaçırmışım.
```
