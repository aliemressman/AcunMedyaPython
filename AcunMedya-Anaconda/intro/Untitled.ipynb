```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

x,y
```

```python
plt.plot(x,y, marker = "s", linestyle = "dotted",color = "red", label = "Satışlar")
plt.legend() # legend -> Bilgi alanı.
plt.xlabel("Günler")
plt.ylabel("Satışlar")
plt.grid()
plt.title("Günlük Satış Tablosu")
plt.show()
```
